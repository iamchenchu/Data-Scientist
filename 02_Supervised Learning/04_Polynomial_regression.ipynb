### Polynomial Regression

```python
# Import Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Assign the data to predictor and outcome variables 
train_data = pd.read_csv('data.csv')
X = train_data['Var_X'].values.reshape(-1, 1)
y = train_data['Var_Y'].values
```

```python
# Create polynomial features
poly_feat = PolynomialFeatures(degree =4)
X_poly = poly_feat.fit_transform(X)
```

```python
# Make the model and fit the model 
poly_model = LinearRegression(fit_intercept=False).fit(X_poly, y)

print(f"Coeffecients : {poly_model.coef_}")
print(f"Intercepts: {poly_model.intercept_}")
```

```python
# Plot the data and polynomial regression model 

X_new = np.linspace(-3, 3.5, 100).reshape(100, 1)
X_new_poly = poly_feat.transform(X_new)
y_new = poly_model.predict(X_new_poly)

plt.scatter(X, y, label="Data")
plt.plot(X_new, y_new, "r-", label="Prediction")
plt.xlabel("Var_X")
plt.ylabel("Var_Y")
plt.legend(loc="best")
plt.show()
```
